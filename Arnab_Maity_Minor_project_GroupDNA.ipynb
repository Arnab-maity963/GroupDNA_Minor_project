```
# Import Required Liabraries

import numpy as np
from datetime import datetime ## Convert timestamp string into datetime objects
```

```
# Load WhatsApp Chat Dataset

with open("hostel_bois.txt","r",encoding="utf-8") as file:
    lines = file.readlines()  ## Read file line by line

print("Total Lines :",len(lines))  ## Total number of lines in dataset
```

```
# Initialize Variables for Analysis

messages = []  ## Store clean messages

participants = set() ## Store unique user

media_count = {} ## Store media message

deleted_count = {}  ## Store deleted message

system_messages = 0 ## Store system messages
```

```
# Chat Data Parsing

for line in lines:  ## Iterate through each line

    line = line.strip() ##  Remove ectra space

    if not line:  ## Skip emepty line
        continue

    try:

        timestamp, rest = line.split(" - ",1)

        if ": " not in rest:
            system_messages += 1
            continue

        sender, message = rest.split(": ",1)

        participants.add(sender)

        if sender not in media_count:
            media_count[sender]=0

        if sender not in deleted_count:
            deleted_count[sender]=0

        if "<Media omitted>" in message: ## Count media meassage
            media_count[sender]+=1

        if "This message was deleted" in message: ## count deelted message
            deleted_count[sender]+=1

        messages.append({
            "timestamp":timestamp,
            "sender":sender,
            "message":message
        })

    except:
        continue
```

```
# Parser Verification & Data Overview

print("Parsed Messages :",len(messages))


print("Participants :",participants)


print("System Messages :",system_messages)


print("Media Count")


print(media_count)


print("Deleted Count")

print(deleted_count)

print("First Five Messages")

for msg in messages[:5]: ## Shows five message for verification
    print(msg)
```

```
# Message per user Analysis

message_count = {}

for msg in messages:

    sender = msg["sender"]  ## Extract sender name

    if sender not in message_count:
        message_count[sender] = 0

    message_count[sender] += 1

print(message_count)
```

```
# Most active user

sorted_users = sorted(
    message_count.items(),
    key=lambda x: x[1],
    reverse=True
)     ## User based message in decending order

print("Message Ranking\n")

for user,count in sorted_users:
    print(f"{user:<10} : {count}")
```

```
# Chat & Date Range Analysis

dates = []  ## Store date of each message

for msg in messages:

    dt = datetime.strptime(
        msg["timestamp"],
        "%d/%m/%y, %H:%M"
    )   ## string timestamp into datetime object

    dates.append(dt)

start_date = min(dates)

end_date = max(dates)

total_days = (end_date-start_date).days + 1

print("Start Date :",start_date)
print("End Date :",end_date)
print("Total Days :",total_days)
```

```
# Group Overview Statistics

print("="*10)
print("GROUP OVERVIEW")
print("="*10)

print("Total Messages :",len(messages))

print("Participants :",len(participants))

print("System Messages :",system_messages)

print("Media Messages :",sum(media_count.values()))

print("Deleted Messages :",sum(deleted_count.values()))

print("Chat Duration :",total_days,"days")
```

```
# MOst & Least active user interaction

most_active = max(
    message_count,
    key=message_count.get
)

least_active = min(
    message_count,
    key=message_count.get
)

print("="*30)

print("MOST ACTIVE USER")

print("="*30)

print(most_active)

print("Messages :",message_count[most_active])

print()

print("="*30)

print("LEAST ACTIVE USER")

print("="*30)

print(least_active)

print("Messages :",message_count[least_active])
```

```
# Active day analysis

day_count = {}

for msg in messages:

    date = msg["timestamp"].split(",")[0]

    if date not in day_count:
        day_count[date] = 0

    day_count[date] += 1

most_day = max(day_count, key=day_count.get)

print("="*30)
print("MOST ACTIVE DAY")
print("="*30)
print("Date :", most_day)
print("Messages :", day_count[most_day])
```

```
# Active hour analysis

hour_count = {}

for msg in messages:

    time = msg["timestamp"].split(", ")[1]

    hour = int(time.split(":")[0])

    if hour not in hour_count:
        hour_count[hour] = 0

    hour_count[hour] += 1

most_hour = max(hour_count,key=hour_count.get)

print("="*15)
print("MOST ACTIVE HOUR")
print("="*15)
print(f"{most_hour}:00 - {most_hour}:59")
print("Messages :",hour_count[most_hour])
```

```
# NumPy Heatmap Initialization

users = sorted(list(participants))

user_index = {}

for i,user in enumerate(users):
    user_index[user] = i

heatmap = np.zeros((len(users),24),dtype=int)

for msg in messages:

    sender = msg["sender"]

    hour = int(
        msg["timestamp"].split(", ")[1].split(":")[0]
    )

    row = user_index[sender]

    heatmap[row][hour] += 1

print(heatmap)
```

```
print("="*80)
print("CHAT ACTIVITY HEATMAP")
print("="*80)

print("Hours →")

for h in range(24):
    print(f"{h:02}", end=" ")

print()

for i, user in enumerate(users):

    print(f"{user:<10}", end=" ")

    for value in heatmap[i]:

        if value == 0:
            symbol = "-" ## no activity

        elif value <= 5:
            symbol = "+" ## low activity

        elif value <= 15:
            symbol = "*" ## Mediam activity

        else:
            symbol = "#" ## high activity

        print(symbol, end=" ")

    print()
```

```
# Night Owl Analysis

print("="*40)
print("NIGHT OWL ANALYSIS")
print("="*40)

for i, user in enumerate(users):

    total = heatmap[i].sum()

    night = (
        heatmap[i][23] +
        heatmap[i][0] +
        heatmap[i][1] +
        heatmap[i][2] +
        heatmap[i][3] +
        heatmap[i][4]
    )

    percentage = (night / total) * 100

    # Activity Level
    if percentage >= 70:
        symbol = "*****"      # Very High
    elif percentage >= 50:
        symbol = "****"       # High
    elif percentage >= 30:
        symbol = "***"        # Medium
    elif percentage >= 10:
        symbol = "**"         # Low
    else:
        symbol = "*"          # Very Low

    print(f"{user:<10} {symbol} {percentage:.2f}%")
```

```
# Text Preprocessing for Word Analysis
## Convert text into lowercase
stop_words = {
    "i","is","the","a","an","and","or","to","of","in","on",
    "for","it","this","that","you","me","my","we","our"
}

word_count = {}

for msg in messages:

    text = msg["message"].lower()

    if "<media omitted>" in text:
        continue

    if "this message was deleted" in text:
        continue

    words = text.split()

    for word in words:

        word = word.strip(".,!?;:'\"()[]{}")

        if word == "":
            continue

        if word in stop_words:
            continue

        if word not in word_count:
            word_count[word] = 0

        word_count[word] += 1

top_words = sorted(word_count.items(), key=lambda x: x[1], reverse=True)

print("="*30)
print("TOP 10 WORDS")
print("="*30)

for word, count in top_words[:10]:
    print(f"{word:<15} : {count}")
```

```
# Top 10 Frequently Used Words

print("="*20)
print("WORD FREQUENCY")
print("="*20)

for word, count in top_words[:10]:

    bar = "#" * (count // 10)

    print(f"{word:<15} {bar} ({count})")
```

```
# Word Frequency Visualization

response_time = {}

for user in users:
    response_time[user] = []

for i in range(1, len(messages)):

    previous = messages[i-1]
    current = messages[i]

    if previous["sender"] != current["sender"]:

        t1 = datetime.strptime(previous["timestamp"], "%d/%m/%y, %H:%M")
        t2 = datetime.strptime(current["timestamp"], "%d/%m/%y, %H:%M")

        minutes = (t2 - t1).total_seconds() / 60

        response_time[current["sender"]].append(minutes)

print("="*30)
print("AVERAGE RESPONSE TIME")
print("="*30)

for user in users:

    if len(response_time[user]) > 0:

        avg = sum(response_time[user]) / len(response_time[user])

        print(f"{user:<10} : {avg:.2f} minutes")
```

```
# Personality Detection of Users

print("="*50)
print("PERSONALITY ANALYSIS")
print("="*50)

for user in users:

    total = message_count[user]

    night = (
        heatmap[user_index[user]][23] +
        heatmap[user_index[user]][0] +
        heatmap[user_index[user]][1] +
        heatmap[user_index[user]][2] +
        heatmap[user_index[user]][3] +
        heatmap[user_index[user]][4]
    )

    percent = (night / total) * 100

    if total == max(message_count.values()):
        personality = "Spammer"

    elif total == min(message_count.values()):
        personality = "Ghost"

    elif percent >= 50:
        personality = "Night Owl"

    else:
        personality = "Regular User"

    print(f"{user:<10} --> {personality}")
```
